## Importing Libraries 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

## step 2 Create the data frame 

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe 
```

##  step 3 - convert the data type of column = rate 

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float (value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

### check missing values 

```python
dataframe.info ()
```

### Type of restaurant 

#### Q.1) what type of restaurant do the majority of customers order from ? 

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel(" type of restaurant")
```

### conclusion of Q1 - majority of the restaurant falls in dinning category 

####  Q2 .) How many votes has each type of restaurant received from customers ?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes' : grouped_data})
plt.plot(result, c= "green", marker ="o")
plt.xlabel("Type of restaurant", c = "red", size = 20)
plt.ylabel("votes", c= "red", size = 20)
```

### conclusion of Q2- dinning restaurant has recieved maximus votes .

##### Which mode recieves maximum ratings 

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

### conclusion of Q5 ) offline (No) order received lower rating in comparison to online(yes) order 

#### Q.6 ) which type of restaurant received more offline orders , so that zomato can prefer   customer with some good offers 

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order' , aggfunc = 'size' , fill_value = 0 )
sns.heatmap(pivot_table, annot = True,cmap = "YlGnBu", fmt = 'd')
plt.title("heatmap")
plt.xlabel("online order")
plt.ylabel("listed_in(type)")
plt.show()
```

### conclusion of Q6) Dining restaurants primarily accept offline orders , whereas cafes primarily receive online orders . This suggests that clients prefer orders in person at resaurants m but perfer online ordering at cafes .
